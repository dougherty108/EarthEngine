```python
# create image collection and see how many images are available
sillybands = ['B2', 'B3', 'B4', 'B5']
landsat_collection = ee.ImageCollection("LANDSAT/LC08/C02/T2").select(sillybands)\
                        .filterBounds(roi)\
                        .filterDate('2017-01-01', '2018-01-01')

print(type(landsat_collection))

print("Number of images available over Taylor Valley Lakes", landsat_collection.size().getInfo())
```

```python
bands = ['B4', 'B3', 'B4', 'B5']
```

```python
def compute_pca(image_collection):
    
    mean = image_collection.select(bands).mean()
    std_dev = image_collection.select(bands).reduce(ee.Reducer.stdDev())
    
    # Compute the principal components
    cov_matrix = image_collection.select(bands).reduce(ee.Reducer.centeredCovariance())
    
    # Compute the principal components
    pca = mean.addBands(std_dev).addBands(cov_matrix)
    
    return pca
```

```python
def extract_endmembers(image_collection):
    pca = compute_pca(image_collection)
    
    # extract pure pixels (endmembers)
    return pca.select(bands).toArray().toArray(1)
```

```python
endmembers = extract_endmembers(landsat_collection)
```

```python
# Define a function for SMA
def apply_sma(image):
    # print bands for debugging
    print('Input bands:', image.bandNames().getInfo())
    
    # Get the image bands
    bands = image.select(sillybands)  # Example with Landsat 8 bands
    # Perform SMA
    sma = bands.unmix(endmembers)
    
    #print bands after SMA for debugging
    print('Bands after SMA:', sma.bandNames().getInfo())
    return image.addBands(sma)
```

```python
# Apply SMA to each image in the collection
sma_collection = landsat_collection.map(apply_sma)

print('type of sma_collection', type(sma_collection))

print("size of sma_collection", sma_collection.size().getInfo())
```

```python
# Export the results to Google Drive
def export_image(image, description):
    task = ee.batch.Export.image.toDrive(
        image=image.select('sma'),
        description='SMA of TV',
        folder='SMA_EarthEngine',
        fileNamePrefix='SMA_',
        scale=30,
        region=roi, 
        crs = 'EPSG:3031',
        maxPixels = 1e9, 
        fileFormat = 'GeoTIFF'
    )
    task.start()
    print('Export task status:', task.status())
```

```python
# Export the first image in the SMA collection as an example
first_image = sma_collection.first()
export_image(sma_collection, "image")

print("Export task started. Check Google Drive for results.")
```
